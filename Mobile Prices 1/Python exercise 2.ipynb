```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
import io

np.seterr(all='raise')

## Task 1
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['mobile_price_1.csv']))
df.head()
```

```python
## 1.3.
inch = 2.54
for i in df['sc_w']:
  if i == np.NaN:
      i = 0
for i in df['sc_h']:
  if i == np.NaN:
      i = 0 
df["resolution"] = np.hypot(df['px_width'], df['px_height']) // (np.hypot(df['sc_h'], df['sc_w'])/ inch)
df.head()
```

```python
## 1.4. 
df["DPI_w"] = (df['px_width'] * inch)// df['sc_w']
print(df)
```

```python
from numpy.core.getlimits import inf
## 1.5.
df["call_ratio"] = df['battery_power']// df['talk_time']
df.head()
# check_nan = df['call_ratio'].isnull().values.any()
# print(check_nan)
```

```python
## 1.6.
df['memory'] = df['memory']// 1000
df.head()
```

```python
## 1.7.
df.describe()
```

```python
## 1.8.
df['speed'] = df['speed'].astype("category")
df['screen'] = df['screen'].astype("category")
df['cores'] = df['cores'].astype("category")
```

```python
## 2.1. 
sum = df['f_camera'].isnull().sum() + df['camera'].isnull().sum()
print(sum)
```

```python
## 2.2.
subAvg = df[(df["sim"] == 'Single') & (df['f_camera'].notnull() | df['camera'].notnull()) & (df['resolution']>12)]
avg = subAvg['battery_power'].sum() / subAvg['battery_power'].count()
print(avg)
```

```python
## 2.3.
subDf = df[(df['wifi'] == 'none') & (df['screen'] == 'Touch') & (df['mobile_wt'] > 145)]
subDf = subDf.sort_values(by=['price'], ascending=False)
print(subDf['id'].iloc[0], subDf['price'].iloc[0])
```

```python
## 2.4.
amount = pd.qcut(df.ram, q=4)
df.bluetooth = pd.Categorical(df.bluetooth, categories=['No', 'Yes'], ordered=False)
df['has_bluetooth'] = np.where(df.bluetooth == 'Yes' ,1 ,0 )
df.pivot_table('has_bluetooth', index = df.gen, columns = amount)
print(df)
```
